## Histogram

```python
iris_df.plot(kind='hist')
```

x-axis shows value range.

y-axis shows frequency.

### histogram of one column

histogram show how many number of data present in the dataset.

```python
iris_df['sepal_length'].plot(kind='hist')
```

Here, 
    between value range of 4.5 to 5.0, there is a frequency of 23 data present.
       between value range of 5.5 to 6.0, there is a frequenct of 15 data present.

```python
iris_df.hist(figsize=(20, 10), color='y')
```

## Boxplots

```python
iris_df.plot(kind='box', figsize=(10, 10))
```

Q1 - Represent 25th percentile

Q2 - Represent 50th percentile.

Q3 - Represent 75th percentile.

Q4 - Represent 100th percentile.

```python
iris_df.describe()
```

IQR = Q3 - Q1

IQR is the difference b/w 75th percentile and 25th percentile.

### IQR for Sepal_Length

```python
IQR = iris_df.describe()['sepal_length']['75%'] - iris_df.describe()['sepal_length']['25%']
```

```python
IQR
```

### IQR for Sepal Width

```python
IQR = iris_df.describe()['sepal_width']['75%'] - iris_df.describe()['sepal_width']['25%']
```

```python
IQR
```

### IQR for Petal Length

```python
IQR = iris_df.describe()['petal_length']['75%'] - iris_df.describe()['petal_length']['25%']
```

```python
IQR
```

### Boxplot in vertical form

```python
iris_df.plot(kind='box', figsize=(10,10), vert=False)
```
