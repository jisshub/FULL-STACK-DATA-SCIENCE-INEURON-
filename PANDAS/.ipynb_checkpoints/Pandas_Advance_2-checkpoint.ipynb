### dropna method

```python
titanic_df.dropna()
```

**Parameter Description:**

axis = 0, means rows

axis = 1, means columns

inplace = True, means permanent updation.

```python
# titanic_df.dropna(inplace=True)
```

```python
# drop columns with nan values, set axis=1
```

```python
# titanic_df.dropna(axis=1)
```

```python
titanic_df.describe()
```

```python
# drop columns where null values > 10
```

```python
titanic_df.dropna(thresh=800, axis=1)
```

**thresh:**

*thresh=800*
- means we set nan vales threshold to 800. means if nan values are more than 
800, we drop that column.

### fillna method

```python
titanic_df.fillna(value='test')
```

```python
# replce nan values with mean value of age
```

```python
titanic_df.fillna(titanic_df['Age'].mean())
```

## Merge Operation

```python
df1 = titanic_df[['Survived', 'Pclass']]
df2 = titanic_df[['Fare', 'Ticket', 'Pclass']]
```

```python
df1.head(3)
```

```python
df2.head(3)
```

```python
pd.merge(df1, df2, on='Pclass', how='left').head()
```

```python
titanic_df.columns
```

```python
df1 = titanic_df[['PassengerId', 'Name']]
df2 = titanic_df[['PassengerId', 'Embarked']]
```

```python
pd.merge(df1, df2, how='right', on='PassengerId').head()
```
