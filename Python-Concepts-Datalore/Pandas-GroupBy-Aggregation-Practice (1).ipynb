```python
import pandas as pd
import numpy as np
```

```python
df_movies = pd.read_csv('netflix_titles.csv')
df_movies
```

```python
df_movies.isna()
```

```python
cleaned_data = df_movies.dropna(axis=0)
cleaned_data
```

1 Write a Pandas program to split the following dataframe into groups based on type.

```python
grouped_df = cleaned_data.groupby('type', as_index=False)
grouped_df.count()
```

2. Write a Pandas program to split the following dataframe by type fetuare
and get mean, min, and max value of duration for each type.

```python
duration_time_list = []
for data in cleaned_data['duration'].str.split():
    data = int(data[0])
    duration_time_list.append(data)
# add duration_time_list as a column to dataframe.
cleaned_data['Duration_Time'] = pd.Series(duration_time_list).values
cleaned_data
# grouped_df.groupby('')
```

```python
cleaned_data.groupby('type').agg(Duration_Mean=('Duration_Time', np.mean), Duration_Max=('Duration_Time', np.max), Duration_Min=('Duration_Time',np.min))
```
