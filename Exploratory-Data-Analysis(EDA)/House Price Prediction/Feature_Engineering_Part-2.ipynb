    Whenever we do real world project, first we have to do train test split - then apply

### 1. Missing Values

```python
# First handle categorical features with nan or missing values
```

```python
features_nan = [feature for feature in house_price_data.columns if house_price_data[feature].isnull().sum() > 1 
                and house_price_data[feature].dtypes == 'O']
for feature in features_nan:
    # take mean of features with nan or null values. 
    # show the percentage of missing values in each feature.     
    print(f'{feature}: {np.round(house_price_data[feature].isnull().mean(), decimals=4)}% missing values')
```

```python
# Replace missing values with label 'Missing'
```

```python
house_price_data[features_nan]= house_price_data[features_nan].fillna('Missing')
```

```python
house_price_data.head()
```

```python
# check for nan values again
```

```python
house_price_data[features_nan].isnull().any()
```

**Inference:**
    
    No more null values in dataframe

```python
# Second handle missing values in numerical columns
```

```python
numerical_nans = [feature for feature in house_price_data.columns if house_price_data[feature].isnull().sum() > 1 
    and house_price_data[feature].dtypes != 'O']
for feature in numerical_nans:
    # take mean of features with nan or null values. 
    # show the percentage of missing values in each feature.    
    print(f'{feature}: {np.round(house_price_data[feature].isnull().mean(), decimals=4)}% missing values')
```

```python
house_price_data[numerical_nans].head()
```

**Replacing missing values in numerical features with median.**

- We know that numerical features have many outliers when we explored the data.
- And whenever we have outlier we should only replace nan or missing values with median/mode and not using mean.

```python
for feature in numerical_nans:
    # find median of each feature     
    median_value = house_price_data[feature].median()
    
    # create new feature (feature_nan) to store nan values
    # if feature value is null, we replace it with 1 else replace it with 0.    
    house_price_data[feature+'_nan'] = np.where(house_price_data[feature].isnull(), 1, 0)
    house_price_data[feature].fillna(median_value, inplace=True)
```

```python
house_price_data.head(50)
```

```python
# Check whether null values are replaced.
```

```python
house_price_data[numerical_nans].isnull().sum()
```

**Inference:**
    
    We replaced all missing values in numerical columns.
    and no null values left.

### Temporal Variables

handling temporal variables

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    # find difference of year in which house was sold and other year features.
    house_price_data[feature]=house_price_data['YrSold']-house_price_data[feature]
```

```python
house_price_data.head()
```

```python
house_price_data[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']]
```

**Inference:**
    
    We got the difference of each year feature and year sold feature.
    new values in each features sepcifies year differences from the year which house was sold
    
  For example:
    
    if YearSold is 2010 and YearBuilt is 1980, then difference is 30. This value 30 shown in 
    updated YearBuilt feature.
    
    if YearSold is 2010 and YearRemodAdd is 2020, then difference is -10. This value -10 shown in 
    updated YearBuilt feature.
    
    Similarly for GarageYrBlt feature.

## Feature Engineering Part 2

In EDA Part we converted some continous features who dont follow guassian distribution to log normal
distribution.

**Features: 'LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice'**


So here we transform only above features who dont follow guassian distribution to log normal distribution.

```python
# Current values
house_price_data[['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].head()
```

```python
for feature in ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']:
    house_price_data[feature] = np.log(house_price_data[feature])
```

```python
# Updated result
house_price_data[['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].head()
```

We took log of each values in the feature.

### Handle rare categorical features

We will remove the categorical features that are present less than 1% of all observations.

```python
# First find categorical feature
```

```python
catgorical_features = [feature for feature in house_price_data.columns if house_price_data[feature].dtypes == 'O']
catgorical_features
```

```python
for feature in catgorical_features:
    
    # find saleprice count by each cateogrical feature.s
    # divide the result by dataset length - will get % value for that specific feature.
    temp = house_price_data.groupby(feature)['SalePrice'].count()/len(house_price_data)
    
    # if % > 0.01, get the index and store it in a new variable.
    temp_df = temp[temp>0.01].index
    
    # if fetuare containf temp_df value then it is replaced with that feature else, 
    # we replace with Rare_Variable label.
    house_price_data[feature] = np.where(house_price_data[feature].isin(temp_df), 
                                         house_price_data[feature], 'Rare_Variable')
    house_price_data
```

### Feature Scaling

- We have many features in the dataset. 
- Each having different measurements and all.
- Also values varies for each feature with other. 
- So it is important to apply feature sclaing.
- So we can apply machine learning algorithm in better way.

```python
# First we have to find on which category we have to perform feature scaling.
```

```python
feature_scale = [feature for feature in house_price_data.columns if feature not in ['Id', 'SalePrice']
                and house_price_data[feature].dtypes != 'O']
house_price_data[feature_scale]
```

So we removed Id and SalePrice column. 

Since id column is not needed, While SalePrice column is our depdendent feature.

#### min-max feature scaling

min max scaler rescales the  data to fixed range of [0, 1].

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# define a scaler
min_max_scl = MinMaxScaler()
```

```python
# fit the scaler paramters
min_max_scl.fit(house_price_data[feature_scale])
```

We applied min max scaler on data

```python
# transform the train and test set - add on the Id and SalePrice variable.
```

```python
data = pd.concat([house_price_data[['Id', 'SalePrice']].reset_index(drop=True),
                 pd.DataFrame(min_max_scl.transform(house_price_data[feature_scale]), 
                                    columns = feature_scale)], axis=1
)
    
```

```python
data.head()
```

**Inference:**
    
    Here we can see that all numerical variables rescaled to the range of (0, 1).
    We can put this scaled data to ML algoithm.
