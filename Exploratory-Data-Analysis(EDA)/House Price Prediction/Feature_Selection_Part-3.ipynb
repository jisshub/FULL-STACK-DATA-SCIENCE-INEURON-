The main aim of the project is to predict the house price based on various features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# for feature selection
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv('./X_train.csv')
```

```python
dataset.head()
```

```python
dataset.tail()
```

<b>Dependent and Independent features<b>

```python
y_train = dataset[['SalePrice']] # dependent variable
y_train.head()
```

```python
X_train = dataset.drop(['Id', 'SalePrice'], axis=1) # independent variable
X_train.head()
```

## Applying Feature Selection

- First Specify the lasso regression model, and select a suitable alpha.
- Bigger the alpha, lesser the features will be selected.
- Use *SelectFromModel* object to select the features whose coefficients are none-zero.

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(X_train, y_train)
```

```python
feature_sel_model.get_support()
```

**Inference:**

Here True basically indicates that the feature is important and should be selected. False indicates that the feature is not important and can be skipped.


<b>Get count of total features & selected features<b>

```python
selected_features = X_train.columns[feature_sel_model.get_support()]
selected_features
```

**Inference:**

    ['MSSubClass', 'LotArea', 'OverallQual', 'YearBuilt', 'YearRemodAdd',
       '1stFlrSF', 'GrLivArea', 'BsmtFullBath', 'Fireplaces', 'GarageCars']
    Above ones are the important features for prediction.


<b>Total features<b>

```python
print(f'Total Features: {X_train.shape[1]}')
print(f'Selected Features: {len(selected_features)}')
print(f'Features with coefficient shrank to zero: {np.sum(feature_sel_model.estimator_.coef_)==0}')
# total_features
```

```python
X_train[selected_features].head()
```

<b>update the X_train with selected features<b>

```python
X_train = X_train[selected_features]
```
