```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla = pdr.get_data_yahoo('TSLA')
df_tesla.head()
```

### Plot the dataset

```python
df_tesla.plot()
```

### Plot on High Column

```python
df_tesla['High'].plot(figsize=(12, 4))
```

### xlimit and ylimit

```python
df_tesla['High'].plot(xlim=['2020-01-01', '2021-01-01'], 
                      ylim=[0,900], 
                      figsize=(12, 4),
                     c='green', linestyle='--')
```

### get data with in a date range

```python
index = df_tesla.loc['2019-01-01':'2021-01-01'].index
```

```python
share_open = df_tesla.loc['2019-01-01':'2021-01-01']['Open']
```

```python
share_open
```

### Draw a subplots

```python
# create figure
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
figure, axis = plt.subplots()

# add more padding tight_layout()
plt.tight_layout()

# prevent any overlapping
figure.autofmt_xdate()

axis.plot(index, share_open)
```
