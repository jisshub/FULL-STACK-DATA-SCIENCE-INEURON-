## 1. Load the libraries

```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

## 2. Select the categorical features from the dataset

```python
# Select the categorical features ['sex','embarked','alone','pclass','Survived']
df=df[['sex','embarked','alone','pclass','survived']]
df.head()
```

## 3. Perform label encoding on each and every categorical feature.

We have to perform label encoding on each categorical features before doing the chi-square test.

### Label Encodning - sex column

```python
from sklearn.preprocessing import LabelEncoder
df['sex'] = LabelEncoder().fit_transform(df['sex'])
df.head()
```

### Label Encodning - alone column

```python
df['alone']=LabelEncoder().fit_transform(df['alone'])
df.head()
```

### Label Encodning - embarked column

```python
df['embarked']=LabelEncoder().fit_transform(df['embarked'])
df.head()
```

## 4. Dependent and Independent features

```python
X=df[['sex', 'embarked', 'alone', 'pclass']]  # independent features 
X.head()
```

```python
y=df[['survived']]
y.head()
```

## 5. Train Test Split the data to avoid overfitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=0)
```

```python
X_train.head()
```

## 6. Perform chi-square on training dataset

**chi-square** test returns two array of below values:
1.   F Scores
2.   P Values

```python
from sklearn.feature_selection import chi2
```

```python
f_score_p_val = chi2(X_train,y_train)

# return 2 arrays having F-Scores and P-Values
f_score_p_val
```

**Inference**

Lesser the p value, more important that feature is.

## 7. Convert P values into series.

```python
p_values = pd.Series(f_score_p_val[1])
p_values
```

```python
p_values.index=X_train.columns
```

```python
p_values
```

## 8. Sort the p values in ascending order

```python
p_values.sort_index(ascending=True)
```

### Observation

Sex Column is the most important column when compared to the output feature Survived.
