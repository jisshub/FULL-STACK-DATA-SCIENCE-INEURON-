# How To Drop Constant Features Using Variance Threshold

Video Link: https://www.youtube.com/watch?v=uMlU2JaiOd8&list=PLZoTAELRMXVPgjwJ8VyRoqmfNs2CJwhVH&index=1

```python
import pandas as pd
```

```python
data=pd.DataFrame({
    'A': [1,2,4,1,3],
    'B': [4,5,6,7,8],
    'C': [0,0,0,0,0],
    'D': [1,1,1,1,1]
})
```

```python
data.head()
```

**Observations:**

Here we can observe values in C and D have Constants features. So we can remove this 2 columns, for that purpose we use a sklearn package called as **Variance Threshold**.

More on: 
https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.VarianceThreshold.html



### Variance Threshold
is a Feature selector that removes all low-variance features. This feature selection algorithm looks only at the features (X), not the desired outputs (y), and can thus be used for unsupervised learning.

**Syntax:**

skearn.feature_selection.VarianceThreshold(threshold=0.0)

If we gave threshold value as 0, then it will go and check each and every feature which has variance as 0. Ie. whichever columns has variance 0, it will be removed.

Here column C has same values and column D has same values. If we compute mean, variance or standard deviation the value will be same as thier column value, because all the values are same.




```python
from sklearn.feature_selection import VarianceThreshold
```

```python
threshold = VarianceThreshold()
threshold.fit(data)
```

```python
threshold.get_support()
```

**Inference:**

Here True represents feature A and B, while False represents feature C and D. Next we pick up the columns C and D that having False value. Because those are the columns needs to be removed. C and D column has 0 variance. They are not important for training.

```python
data.columns[threshold.get_support()] 
```

<b>Get the columns that need to be removed<b>

```python
constant_columns = [column for column in data.columns if column not in data.columns[threshold.get_support()]]
constant_columns
```

<b>Drop the constant columns<b>

```python
data.drop(constant_columns, axis=1)
```

## Practice on Bigger Dataset

Dataset Source: https://www.kaggle.com/competitions/santander-customer-satisfaction/data?select=train.csv



### 1. Load the dataset

```python
df = pd.read_csv('santander.csv', nrows=10000)
df.head()
```

```python
df.shape
```

### 2. Dependent and Independent features

```python
X=df.drop(['TARGET'], axis=1) # independent features
X.head()
```

```python
y=df[['TARGET']] # dependent feature
y.head()
```

### 3. Split the dataset into training and test set.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=0)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
y_test.head()
```

### 4. Apply Variance threshold feature selector

```python
var_thresh = VarianceThreshold()
var_thresh.fit(X_train)
```

### 5. Find non constant feature

```python
var_thresh.get_support()
```

```python
sum(var_thresh.get_support())
```

**Inference**

Total of 284 feature are not constant. So out of 370 features, 284 are not constant and remaining 86 are constant features.

### 6. Get the constant columns

```python
constant_features = [column for column in X_train.columns if column not in X_train.columns[var_thresh.get_support()]]
# [column for column in data.columns if column not in data.columns[threshold.get_support()]]
constant_features
```

**Inference**

These are the columns having zero variance.

### 7. Drop the constant features from training set

```python
X_train.drop(constant_features, axis=1, inplace=True)
```

```python
X_train.shape
```

**Inference**

After all the filtering, we have 284 features in training set
