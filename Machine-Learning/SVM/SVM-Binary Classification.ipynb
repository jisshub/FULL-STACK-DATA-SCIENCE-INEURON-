```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.pandas.set_option('display.max_columns', None)
```

## Load dataset

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df=df[df['species']!='setosa']
```

```python
df.head()
```

```python
df['species'].unique()
```

## Assign labels to species values.

```python
df['species']=df['species'].map({"versicolor":0,"virginica":1})
```

```python
df.head()
```

## Divide dataset into independent and dependent features

```python
y = df.iloc[:, :-1] # independent features
X = df.iloc[:, -1]  # dependent feature
```

```python
X.head()
```

```python
y.head()
```

## train test split the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.head()
```

## write sklearn logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
regression = LogisticRegression()
```

```python
regression
```

## find alpha value using gridsearch cv

```python
from sklearn.model_selection import GridSearchCV

parameter={'C':[0.5,1,2,3,5,6,7,9,10,30,45],'penalty':['l2','elasticnet']}
log_regressor=GridSearchCV(regression,parameter,scoring='accuracy',cv=5)

log_regressor.fit(X,y)
```
