## Extract the date of journey intp month year and date

```python
dataset['Date']= dataset['Date_of_Journey'].str.split('/').str[0]
dataset['Month'] = dataset['Date_of_Journey'].str.split('/').str[1]
dataset['Year'] = dataset['Date_of_Journey'].str.split('/').str[1]
```

```python
dataset.head()
```

```python
dataset.info()
```

## convert date month year type to integer

```python
dataset.Date = dataset['Date'].astype(int)
dataset.Month = dataset['Month'].astype(int)
dataset.Year = dataset['Year'].astype(int)
```

```python
dataset.info()
```

## remove date_of_journey columns now

```python
dataset.drop(columns='Date_of_Journey', axis=1, inplace=True)
```

## Dealing with arrival times feature

```python
dataset.Arrival_Time
```

### Extract the time from Arrival_Time column

```python
dataset['Arrival_time'] = dataset['Arrival_Time'].str.split(' ').str[0]
dataset.Arrival_time
```

### Extract arrival hour from time

```python
dataset['Arrival_Hour'] = dataset['Arrival_Time'].str.split(':').str[0]
```

```python
dataset.Arrival_Hour
```

### Extract  Arrival Minutes from Hour

```python
dataset['Arrival_Minute'] = dataset['Arrival_Time'].str.split(':').str[1]
```

```python
dataset.Arrival_Minute
```

### Remove Arrival_Time Column

```python
dataset.drop('Arrival_Time', axis=1, inplace=True)
```

## One hot encoding the categorical variables

- Creating new dummy columns

```python
encoded = pd.get_dummies(dataset, columns=['Source', 'Destination', 
                                         'Additional_Info', 'Airline'], 
                                drop_first=True)
encoded.head()
```

## Drop the encoded columns

```python
dataset.drop(columns=['Source', 'Destination', 
                    'Additional_Info', 'Airline'], axis=1, inplace=True)
```

## Concat new dummy columns to dataframe.

```python
dataset = pd.concat([dataset, encoded], axis=1)
dataset.head()
```
