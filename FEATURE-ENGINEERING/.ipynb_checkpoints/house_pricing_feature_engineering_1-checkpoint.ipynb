## Handle missing values

```python
nan_features = [feature for feature in df.columns if df[feature].isnull().sum()>=1]
```

```python
for feature in nan_features:
       print(f'''{feature}: {np.round(df[feature].isnull().mean(),
                        decimals=4)}% missing values''')
```

```python
df.isnull().sum()
```

## Working with categorical features

```python
categorical_features_nan = [feature for feature in df.columns if df[feature].isnull().sum()>=1 
                                        and  df[feature].dtypes=='O']
categorical_features_nan
```

## Replacing missing categorical values

missing values are imputed with mode of the feature.

```python
for feature in categorical_features_nan:
    df[feature].fillna(df[feature].mode()[0], inplace=True)
```

```python
df.head()
```

## Working with numerical features

```python
numerical_features_nan = [feature for feature in df.columns if df[feature].isnull().sum()>=1 
                                        and  df[feature].dtypes!='O']
numerical_features_nan
```

## Replacing missing numerical features

```python
for feature in numerical_features_nan:
    # find median of each numerical nan values.
    median_values = df[feature].median()
    # create new column for null and and non null values.
    # in that column, nan is replaced with 1 and non null replaced with 0     
    df[feature+'nan'] = np.where(df[feature].isnull(),1,0)
    # finally fill the feature with median values.    
    df[feature].fillna(median_values, inplace=True)
```
